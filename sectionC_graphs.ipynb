```python
import pandas as pd
df = pd.read_csv("df.csv",header=None)
```

```python
col=df.iloc[0,1:]
index=df.iloc[1:,0]
df=df.iloc[1:,1:]
df.columns=col
df.set_index(index,inplace=True)
```

```python
df['industry'].isnull().values.any()   #False means no NaN value
```

```python
import re
def ind_catagorize(x):  
    if (re.search(r'^Engineering',x)):
        return str('Engineering')
    else:
        return x
df['Industry']=df['industry'].apply(ind_catagorize)
df.head()
```

```python
df['Industry'].value_counts()
```

```python
Top_15_industries=df['Industry'].value_counts()[0:16]
```

```python
sns.barplot(x=Top_15_industries,y=Top_15_industries.index)
```

```python
df['job title'].value_counts()[0:16]
```

```python
def job_catagorize(x):  
    if (re.search('Business Analyst',x)):
        return str('Business Analyst')
    elif (re.search('Data Analyst',x)):
        return str('Data Analyst')
    elif (re.search('Data Scientist',x)):
        return str('Data Scientist')
    elif (re.search('Data Engineer',x)):
        return str('Data Engineer')
    else:
        return x
df['Job title']=df['job title'].apply(job_catagorize)
df.head()
```

```python
df['Job title'].value_counts()[0:16]
```

```python
#Business Analyst
def findindustry(i):
    k=df['Job title']==i
    return df[k]["Industry"]

result=list(map(findindustry,df['Job title'].value_counts()[0:16].index))

top10ind0=result[0].value_counts()[0:10]
top10ind0
```

```python
sns.barplot(x=top10ind0,y=top10ind0.index).set_title('Business Analyst',fontsize=20)
```
