```python
import pandas as pd
dfk = pd.read_csv("df_kent.csv",header=None)
dfe1 = pd.read_csv("df_eddie_1.csv",header=None)
dfe2 = pd.read_csv("df_eddie_2.csv",header=None)
dfi = pd.read_csv("df_ian.csv",header=None)
dfd = pd.read_csv("df_daniel.csv",header=None)
df269 = pd.read_csv("df269.csv",header=None)
```

```python
chopdf=list(map(lambda x:x.iloc[1:,:],[df269,dfe2,dfd,dfi,dfk]))
```

```python
from functools import reduce
df=reduce(lambda x,y:x.append(y),[dfe1]+chopdf)
```

```python
df
```

```python
col=df.iloc[0,1:]
index=df.iloc[1:,0]
df=df.iloc[1:,1:]
df.columns=col
df.set_index(index,inplace=True)
```

```python
df
```

```python
df.to_csv(path_or_buf="df.csv")
```
